```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array(np.arange(-10,10,0.2))
```

```python
#identity
plt.plot(x,x)
plt.show()
```

```python
#sigmoid
y=1/(1+np.exp(-x))
plt.plot(x,y)
plt.plot(x,y*(1-y))
plt.show()
```

```python
#tanh
y=np.tanh(x)
plt.plot(x,y)
plt.plot(x,1-np.square(y))
```

```python
#arctan
y=np.arctan(x)
plt.plot(x,y)
plt.plot(x,1/(1+np.square(x)))
```

```python
#softmax
y=np.log(1+np.exp(-x))
plt.plot(x,y)
plt.plot(x,1/(1+np.exp(-x)))
plt.show()
```

```python
#LeakyRelu
lst=[]
der=[]
for i in x:
    if i<=0:
        lst.append(0.01*i)
        der.append(0.01)
    else:
        lst.append(i)
        der.append(1)
lst=np.array(lst)
plt.plot(x,lst)
plt.plot(x,der)
plt.show()
```

```python
#relu
lst=[]
der=[]
for i in x:
    if i<0:
        lst.append(0)
        der.append(0)
    else:
        lst.append(i)
        der.append(1)
plt.plot(x,lst)
plt.plot(x,der)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,2*np.pi,400)
y=np.sin(x**2)
```

```python
fig,(ax1,ax2)=plt.subplots(1,2)
ax1.plot(x,y)
ax2.plot(x,-y)
```

```python
fig,axs=plt.subplots(2,2)
axs[0,0].plot(x,y,"tab:blue")
axs[0,0].plot("Axis [0,0]")
axs[0,1].plot(x,y,"tab:orange")
axs[0,1].plot("Axis [0,1]")
axs[1,0].plot(x,-y,"tab:green")
axs[1,0].plot("Axis [1,0]")
axs[1,1].plot(x,-y,"tab:red")
axs[1,1].plot("Axis [1,1]")
```

```python
fig,(ax1,ax2)=plt.subplots(2)
fig.suptitle('Axes values')
ax1.plot(x,y)
ax2.plot(x+1,-y)
```
