```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x,y=make_blobs(n_samples=300,n_features=2,centers=3,cluster_std=0.5,shuffle=True,random_state=0)
```

```python
plt.scatter(x[:,0],x[:,1],c='white',edgecolor='black',marker='o',s=50)
plt.grid()
plt.show()
```

```python
km=KMeans(n_clusters=3,init='random',n_init=10,max_iter=300,random_state=0)
```

```python
y_km=km.fit_predict(x)
```

```python
center=km.cluster_centers_
```

```python
center
```

```python
plt.figure()
plt.scatter(x[:,0],x[:,1],s=50,c='lightgreen',marker='d',edgecolor='black')
plt.scatter(center[0,0],center[0,1],s=80,c='orange',marker='o',edgecolor='black',label='Center 1')
plt.scatter(center[1,0],center[1,1],s=80,c='blue',marker='>',edgecolor='black',label='Center 2')
plt.scatter(center[2,0],center[2,1],s=80,c='red',marker='*',edgecolor='black',label='Center 3')
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
print("Distortion={}".format(km.inertia_))
```

```python
#elbow method
distortions=[]
for i in range(2,11):
    km=KMeans(n_clusters=i,init='k-means++',n_init=10,max_iter=300,random_state=0)
    km.fit(x)
    distortions.append(km.inertia_)
```

```python
plt.plot(range(2,11),distortions,marker='o')
plt.xlabel('No of CLusters')
plt.ylabel('Distortions')
plt.show()
```
