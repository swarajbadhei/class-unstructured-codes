```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
X=np.random.random_integers(low=200,high=500,size=(10,1))
y=np.random.random_integers(low=100,high=300,size=(10))
```

```python
lr=LinearRegression()
pr=LinearRegression()
```

```python
quadratic=PolynomialFeatures(degree=2)
```

```python
X_quad=quadratic.fit_transform(X)
```

```python
print(X_quad.shape)
print(y.shape)
```

```python
lr.fit(X_quad,y)
```

```python
x_fit=np.random.random_integers(low=200,high=500,size=(10,1))
y_quad_fit=lr.predict(quadratic.fit_transform(x_fit))
```

```python
plt.plot(x_fit,y_quad_fit,label='linear fit',linestyle='--')
```
