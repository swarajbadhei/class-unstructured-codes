```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris=datasets.load_iris()
```

```python
target=iris.target
```

```python
iris=iris.data
```

```python
for i in range(3):
    for j in range(3):
        plt.scatter(iris[:,i],iris[:,j],c=target)
        plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris, target, test_size=0.33, random_state=42)
```

```python
pep=Perceptron(eta0=0.2,penalty='l1')
```

```python
pep.fit(X_train,y_train)
```

```python
pred=pep.predict(X_test)
```

```python
accuracy_score(pred,y_test)
```

```python
for i in range(1,4):
    X_train, X_test, y_train, y_test = train_test_split(iris[:,:1], target, test_size=0.33, random_state=42)
    pep=Perceptron(eta0=0.3,penalty='l1')
    pep.fit(X_train,y_train)
    pred=pep.predict(X_test)
    print(accuracy_score(pred,y_test))
    print("it is from 0 to "+str(i))
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(iris[:,:3], target, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```
