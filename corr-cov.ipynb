```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,50,0.5)
y=np.sin(x)
```

```python
plt.plot(x,y)
```

```python
co_relation=np.sum((x-np.mean(x))*(y-np.mean(y)))/(np.sqrt(np.sum(np.square((x-np.mean(x)))))*np.sqrt(np.sum(np.square((y-np.mean(y))))))
```

```python
co_relation
```

```python
np.corrcoef(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x=np.arange(0,10,0.5)
y=x*0.9+np.random.random_sample((x.shape))
```

```python
plt.scatter(x,y)
```

```python
lr=LinearRegression().fit(x.reshape(-1,1),y)
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x.reshape(-1,1)))
```

```python
mse=np.mean(np.sum(np.square(y-lr.predict(x.reshape(-1,1)))))
```

```python
r_sq=1-(mse/np.var(y))
```

```python
print("mse={} \nr_squre={}".format(mse,r_sq))
```
